```python

# Lagrange Interpolation Formula
def lagrange_interpolation(x, y, xp):
    n = len(x)
    yp = 0
    for i in range(n):
        p = 1
        for j in range(n):
            if j != i:
                p *= (xp - x[j]) / (x[i] - x[j])
        yp += y[i] * p
    return yp

# Input values
x = list(map(float, input("Enter the values of x (space-separated): ").split()))
y = list(map(float, input("Enter the values of f(x) (space-separated): ").split()))

if len(x) != len(y):
    print("Error: x and f(x) must have the same number of elements.")
else:
    xp = float(input("Enter the value of x for interpolation: "))
    yp = lagrange_interpolation(x, y, xp)
    print(f"For x = {xp}, interpolated value of y = {yp}")
```

```python
class Account:
    def __init__(self,acc_no,name,acc_type,balance):
        self.acc_no = acc_no
        self.name = name
        self.acc_type = acc_type
        self.balance = balance
    def calculate_balance(self):
        if self.acc_type == 'current':
            return self.balance  # No interest for current accounts
        elif self.acc_type == 'saving':
            return self.balance + (self.balance * 0.0270)
        elif self.acc_type == 'fixed_deposit':
            return self.balance + (self.balance * 0.070)
        else:
            raise ValueError("Invalid account type")

    def __str__(self):
        calculated_balance = self.calculate_balance()
        return (f"Account No: {self.acc_no}\n"
                f"Name: {self.name}\n"
                f"Account Type: {self.acc_type}\n"
                f"Balance (with interest): {calculated_balance:.2f}")
# Example Accounts
account1 = Account(101, "yash", "saving", 5000)
account2 = Account(102, "dash", "current", 10000)
account3 = Account(103, "raja", "fixed_deposit", 15000)

# Print details of each account
print(account1)
print()
print(account2)
print()
print(account3)
```

```python
class FamilyMember:
    def __init__(self, name, age, role, contribution=0):
        self.name = name
        self.age = age
        self.role = role
        self.contribution = contribution

    def __str__(self):
        return (f"Name: {self.name}\n"
                f"Age: {self.age}\n"
                f"Role: {self.role}\n"
                f"Monthly Contribution: {self.contribution:.2f}")


class FamilyBudget:
    def __init__(self, total_income, expenses):
      
        self.total_income = total_income
        self.expenses = expenses

    def calculate_savings(self):
        total_expenses = sum(self.expenses.values())
        return self.total_income - total_expenses

    def add_expense(self, category, amount):
        self.expenses[category] = amount

    def __str__(self):
        expenses_str = "\n".join([f"{key}: {value:.2f}" for key, value in self.expenses.items()])
        savings = self.calculate_savings()
        return (f"Total Income: {self.total_income:.2f}\n"
                f"Expenses:\n{expenses_str}\n"
                f"Monthly Savings: {savings:.2f}")


# Example Usage
# Family Members
member1 = FamilyMember("yash", 40, "Parent", 2000)
member2 = FamilyMember("dash", 42, "Parent", 3000)
member3 = FamilyMember("raja", 18, "Child", 500)

# Print family members
print(member1)
print(member2)
print(member3)

# Family Budget
expenses = {
    "rent": 1500,
    "groceries": 800,
    "utilities": 300,
    "education": 600,
}
total_income = member1.contribution + member2.contribution + member3.contribution

family_budget = FamilyBudget(total_income, expenses)
print("\nFamily Budget:")
print(family_budget)

# Adding a new expense
family_budget.add_expense("transportation", 200)
print("\nUpdated Family Budget:")
print(family_budget)
```
